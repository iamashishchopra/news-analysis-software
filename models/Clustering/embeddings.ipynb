```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
df.head()
```

```python
embedder = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
embeddings = embedder.encode(df["Heading"].values)
print(embeddings)
```

```python
np.save('embeddings_headings.npy', embeddings)
```
